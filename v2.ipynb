```python
%%bash
# download sentence pairs
wget "http://www.manythings.org/anki/cmn-eng.zip" -O cmn-eng.zip 2>&1 | grep ERROR
unzip -oq cmn-eng.zip -d cmn-eng && echo "done"
```

```python
# words and characters which shouldn't be treated as words
ignored = {"？", "?", "：", "－", "，", ",", "）","（", "％", "＂", "！", "!", "﹐", "。", ".", "》", "《", "·", "“", "”", "\"", " ", "​"}
```

```python
# load sentences and calculate word frequency

import jieba
from hanziconv import HanziConv

snts_dict = {}
words_dict = {}

with open("cmn-eng/cmn.txt") as f:
    lines = f.readlines()
    
for line in lines:
    fields = line[:-1].split("\t")
    if not fields[1] in snts_dict:
        words = jieba.lcut(fields[1])
        wc = 0 # word count, excluding ignored
        for word in set(words):
            if word.isdigit() or word.isascii():
                ignored.add(word)
            if not word in ignored:
                wc += 1
                if not word in words_dict:
                    words_dict[word] = {
                        "cmn": word,
                        "freq": 0, # number of occurences of this word in the dataset
                        "snts": [] # indices of sentences which contain this word
                    }
                words_dict[word]["freq"] += 1
        snts_dict[fields[1]] = {
            "cmn": fields[1],
            "words": words,
            "unknown": wc, # number of words which are not learned yet
            "translations": []
        }
    snts_dict[fields[1]]["translations"].append({"eng": fields[0], "attr": fields[2]})

ready = set() # indices of sentences which contain exactly one unknown word
known = set() # words which are known
    
snts = [] # list of sentences
count = 0
for snt in snts_dict.values():
    for word in set(snt["words"]):
        if not word in ignored:
            words_dict[word]["snts"].append(count)
    if snt["unknown"] == 1:
        ready.add(count)
    snts.append(snt)
    count += 1

# words = sorted(words_dict, reverse=True, key=(lambda w : words_dict[w]["freq"])) # most common words are at the beginning    
# words = list(map(lambda w : (words_dict[w]["freq"], words_dict[w]["cmn"]), words))
# count1 = 0
# for i in range(len(words)):
#     if words[i][0] == 1:
#         count1 += 1
# print(count1)

# construct deck

deck = []
maxl = 0 # if not 0, then it is the maximal amount of cards in the deck
c = 0
skipped = 0
unskipped = 0
while len(ready) != 0:
    c += 1
    if maxl != 0 and c > maxl:
        break
    best_freq = 0
    best_word = ""
    best_sid = 0
    for sid in ready:
        for word in set(snts[sid]["words"]):
            if (not word in ignored) and (not word in known) and words_dict[word]["freq"] >= best_freq: # TODO: maybe a more sophisticated algorithm to choose the best sentence
                best_freq = words_dict[word]["freq"]
                best_word = word
                best_sid = sid
    if best_word == "":
        print("Couldn't find next word to use, ready:", ready)
        break
    assert best_word != ""
    deck.append({
        "word": words_dict[best_word]["cmn"],
        "freq": words_dict[best_word]["freq"],
        "snt": snts[best_sid]["cmn"],
        "snt_words": snts[best_sid]["words"],
        "snt_translations": snts[best_sid]["translations"]
    })
    known.add(best_word)
    for sid in set(words_dict[best_word]["snts"]):
        snts[sid]["unknown"] -= 1
        if snts[sid]["unknown"] == 1:
            ready.add(sid)
        elif snts[sid]["unknown"] == 0:
            ready.remove(sid)
        assert snts[sid]["unknown"] >= 0
    while len(ready) == 0 and len(words_dict) - len(known) != 0:
        # find the smallest amount of unknown words in a sentence (>=2)
        unk = 1000
        for snt in snts:
            assert snt["unknown"] != 1, snt
            if snt["unknown"] > 0 and snt["unknown"] < unk:
                unk = snt["unknown"]
                if unk == 2:
                    break
        # find the most frequent word among remaining sentences with `unk` unknown words
        freq = {}
        for snt in snts:
            if snt["unknown"] == unk:
                for word in set(snt["words"]):
                    if not word in ignored and not word in known:
                        if not word in freq:
                            freq[word] = 0
                        freq[word] += 1
        if len(freq) == 0:
            print("freq is empty, unk =", unk, "len(words_dict) =", len(words_dict), "len(known) =", len(known), "len(ready) =", len(ready))
            unks = {}
            for snt in snts:
                unk = snt["unknown"]
                if not unk in unks:
                    unks[unk] = 0
                unks[unk] += 1
            print(unks)
            break
        assert len(freq) > 0
        word = sorted(freq, reverse=True, key=freq.get)[0]
        known.add(word)
        skipped += 1
        added = False
        for sid in set(words_dict[word]["snts"]):
            snts[sid]["unknown"] -= 1
            if snts[sid]["unknown"] == 1:
                ready.add(sid)
                # normally this word would be skipped, but we allow it to be added with an i+2 sentence (where 2 words are unknown)
                # when we know that the second unknown word in that sentence will be in one of the sentences in `ready`
                if not added:
                    deck.append({
                        "word": words_dict[word]["cmn"],
                        "freq": words_dict[word]["freq"],
                        "snt": snts[sid]["cmn"],
                        "snt_words": snts[sid]["words"],
                        "snt_translations": snts[sid]["translations"]
                    })
                    unskipped += 1
                    added = True
            assert snts[sid]["unknown"] >= 1, (snts[sid], unk)
            
for i in range(len(deck)):
    deck[i]["def_link"] = "https://en.m.wiktionary.org/wiki/" + HanziConv.toTraditional(deck[i]["word"]) + "#Definitions"

print("finished,", len(words_dict) - len(known), "words not reached;", skipped - unskipped, "skipped out of", skipped, "non i+1 sentences")
```

```python
with open("deck.tsv", "w") as f:
    f.write("#separator:Tab\n#html:true\n#columns:word\tsentence\ttranslation\tadditional translations\twiktionary link\tattributions\n")
    for n in deck:
        for i in range(len(n["snt_words"])):
            if n["snt_words"][i] == n["word"]:
                n["snt_words"][i] = "<b>" + n["snt_words"][i] + "</b>"
        snt = "".join(n["snt_words"])
        word = n["word"]
        transl = n["snt_translations"][0]["eng"]
        transl_pl = []
        if len(n["snt_translations"]) > 1:
            transl_pl = [t["eng"] for t in n["snt_translations"][1:]]
        transl_pl = "<br>".join(transl_pl)
        attr = "<br>".join([t["attr"] for t in n["snt_translations"]])
        link = n["def_link"]
        f.write(f"{word}\t{snt}\t{transl}\t{transl_pl}\t{link}\t{attr}\n")
```

```python
import matplotlib.pyplot as plt

xs = []
ys = []

total = 0
for word in words_dict.values():
    total += word["freq"]
    
fig, ax = plt.subplots()
    
acc = 0
eighty = False
ninety = False
for i in range(len(deck)):
    acc += words_dict[deck[i]["word"]]["freq"]
    xs.append(i)
    ys.append(100*acc/total)
    if not eighty and 100*acc/total >= 80:
        ax.axvline(i, label=i)
        ax.text(i+50,2,i)
        ax.text(i+100,7,"80%")
        eighty = True
    elif not ninety and 100*acc/total >= 90:
        ax.axvline(i, label=i)
        ax.text(i+50,2,i)
        ax.text(i+100,7,"90%")
        ninety = True
    
ax.plot(xs, ys)
ax.grid(True)
fig.set_figwidth(15)
plt.show()
```
